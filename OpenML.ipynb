```python
openml_list = oml.datasets.list_datasets()
datalist = pd.DataFrame.from_dict(openml_list, orient='index') 
datalist = datalist[[
    'did','name','NumberOfInstances',
    'NumberOfFeatures','NumberOfClasses'
]]
```

```python
datasets = datalist[
    (datalist['NumberOfInstances'] > 1000) & 
    (datalist['NumberOfInstances'] < 2000) &
    (datalist['NumberOfFeatures'] < 100) &
    (datalist['NumberOfClasses'] > 0) &
    (datalist['NumberOfClasses'] < 10) 
].sample(10)
datasets
```
